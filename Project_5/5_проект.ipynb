### Обработаем пропуски при необходимости:

Создадим функцию для поиска строк с пропусками

```python
def show_empty_data(col):
    return games[games[col].isnull()]
```

Об этих играх на платформе GEN мало данных и их всего 2. Предлагаю убрать их из данных.

В исходном датафрейме всего 16715 строк, но при этом 11560 уникальных наименований игр. Это означает что, одна и та же игра могла выпуститься на разных платформах. Следовательно, пропущенные значения в датах релиза могут принадлежать одним и тем же играм. Обычно игровые студии стараются выпускать в релиз игры на все платформы одновременно, значит мы можем использовать даты релиза других платформ для заполнения пропусков.

```python
filled_year = games[pd.notnull(games['year_of_release'])]
```

```python
def fill_year(row, **col):
    rows = filled_year[filled_year['name'] == row['name']]

    if len(rows) > 0:
        return rows.iloc[0]['year_of_release']
    return row['year_of_release']
```

```python
games['year_of_release'] = games.apply(fill_year, axis=1)
```

```python
games['year_of_release'].isnull().sum()
```

После проведения замены, 146 игр так и остались с неизвестной датой релиза, но относительно всего датасета 0,87% - это небольшая часть, пропуски можно удалить

```python
games = games[pd.notnull(games['year_of_release'])]
```

Соотнесем выводы по PS4 с продажами игр на других платформах.

```python
relevant_platforms = ["3DS", "PSV", "WiiU", "XOne", "PC", "X360", "PS3"]
```

```python
# используем цикл для построения диаграмм рассеяния
# диаграммы зависимости продаж от оценок критиков и пользователей расположим рядом слева и справа соответственно

for name_platform in relevant_platforms:
    x_critic = relevant_games[relevant_games['platform']==name_platform]['critic_score']
    y_critic = relevant_games[relevant_games['platform']==name_platform]['total_sales']
    
    x_user = relevant_games[relevant_games['platform']==name_platform]['user_score']
    y_user = relevant_games[relevant_games['platform']==name_platform]['total_sales']
    
    
    fig = plt.figure(figsize=(12, 5))
    
    ax1 = fig.add_subplot(121)
    ax1.scatter(x_critic, y_critic)
    ax1.set_title('Зависимость продаж от оценок критиков\n на платформе '+name_platform, fontsize=14, fontweight="bold")
    ax1.set_xlabel('Оценки критиков\n', fontsize=14)
    ax1.set_ylabel('Продажи', fontsize=14)
    
    ax2 = fig.add_subplot(122)
    ax2.scatter(x_user, y_user)
    ax2.set_title('Зависимость продаж от оценок пользователей\n на платформе '+name_platform, fontsize=14, fontweight="bold")
    ax2.set_xlabel('Оценки пользователей\n', fontsize=14)
    ax2.set_ylabel('Продажи', fontsize=14)
    
    plt.tight_layout(w_pad=3)
```

```python
all = relevant_games[relevant_games['platform'] != 'PS4']
columns = ['user_score', 'critic_score', 'total_sales']
```

```python
all[columns].corr()
```

На основании данног исследования определили:
- С увеличением среднего рейтинга критиков увеличивается и средний уровень выручки с продажи игр, чем качественнее игра, тем больше денег она приносит разработчику и издателю.
- С увеличением среднего рейтинга игроков увеличивается и средний уровень выручки с продажи игр, правда так называемых выбросов больше - когда рейтинг пользователей не велик, но тем не менее игра собирает большие деньги. К примеру все покупают игру, принося разработчикам прибыль, а потом расстраиваются, увидев, какая эта игра, и ставят ей довольное средние оценки.

В целом, результаты очень похожи, как и в выбранной отдельно PS4, оценки критиков оказывают чуть большее влияние на продажи, чем оценки пользователей.

## Шаг 5. Проверьте гипотезы

### 1 Гипотеза. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

В данной проверке гипотезы, мы рассматриваем две генеральные совокупности - игры для платформ PC и XBox

Нулевая гипотеза H₀: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Альтернативная гипотеза H₁: Средние пользовательские рейтинги платформ Xbox One и PC различаются

```python
pc = relevant_games[relevant_games['platform'] == 'PC'] 
xone = relevant_games[relevant_games['platform'] == 'XOne']

alpha = .01
results = st.ttest_ind(pc['user_score'], xone['user_score'], nan_policy='omit')
print('значение p-value: ', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем Нулевую гипотезу')
else:
    print('Не получилось отвергнуть Нулевую гипотезу')
```

### 2 Гипотеза. Средние пользовательские рейтинги жанров Action и Sports разные

В данной проверке гипотезы, мы рассматриваем две генеральные совокупности - игры жанров Action и Sports

Нулевая гипотеза H₀: Средние пользовательские рейтинги жанров Action и Sports одинаковые

Альтернативная гипотеза H₁: Средние пользовательские рейтинги жанров Action и Sports различаются

```python
action = relevant_games[relevant_games['genre'] == 'Action'] 
sports = relevant_games[relevant_games['genre'] == 'Sports']

alpha = .05
results = st.ttest_ind(action['user_score'], sports['user_score'], nan_policy='omit')
print('значение p-value: ', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем Нулевую гипотезу')
else:
    print('Не получилось отвергнуть Нулевую гипотезу')
```

```python
action['user_score'].mean()
```

```python
action['user_score'].median()
```

Получив p-value < alpha, мы отвергли Нулевую гипотезу. Можем говорить что среднии пользовательские рейтинги жанров Action и Sports различаются.
